```python
cout = 0
# ticker in top 60 etfs in 2018 not in top 60 etfs in 2011 
changed = []
same = []
for i in list(etf_2018['Symbol']):
    if i in list(etf_2011['Ticker']):
        cout +=1
        same.append(i)
    else:
        changed.append(i)
        
        
```

```python
# ticker in top 60 etfs in 2011 not in top 60 etfs in 2018
changed2011 = set(etf_2011['Ticker']) - set(same)
```

```python
xlf = pd.read_csv('XLF.csv')
xle = pd.read_csv('XLE.csv')
oih = pd.read_csv('OIH.csv')
xlk = pd.read_csv('XLK.csv')
ige = pd.read_csv('IGE.csv')
pho = pd.read_csv('PHO.csv')
xlp = pd.read_csv('XLP.csv')
#xlv = pd.read_csv('assignment7/XLV.csv')
xlu = pd.read_csv('XLU.csv')
moo = pd.read_csv('MOO.csv')
gdx = pd.read_csv('GDX.csv')
pph = pd.read_csv('PPH.csv')
pbw = pd.read_csv('PBW.csv')
iyr = pd.read_csv('IYR.csv')
#bbh = pd.read_csv('assignment7/BBH.csv')
```

```python
xlf_close = xlf['Adj Close']
xle_close = xle['Adj Close']
oih_close = oih['Adj Close']
xlk_close = xlk['Adj Close']
ige_close = ige['Adj Close']
pho_close = pho['Adj Close']
xlp_close = xlp['Adj Close']
#xlv_close = xlv['Adj Close']
xlu_close = xlu['Adj Close']
moo_close = moo['Adj Close']
gdx_close = gdx['Adj Close']
pph_close = pph['Adj Close']
pbw_close = pbw['Adj Close']
iyr_close = iyr['Adj Close']
#bbh_close = bbh['Adj Close']
```

```python
df = pd.DataFrame()
df['Date'] = pd.read_csv('MOO.csv')['Date']
```

```python
df['xlf_close'] = xlf_close
df['xle_close'] = xle_close
df['oih_close'] = oih_close
df['xlk_close'] = xlk_close
df['ige_close'] = ige_close
df['pho_close'] = pho_close
df['xlp_close'] = xlp_close
#df['xlv_close'] = xlv_close
df['xlu_close'] = xlu_close
df['moo_close'] = moo_close
df['gdx_close'] = gdx_close
df['pph_close'] = pph_close
df['pbw_close'] = pbw_close
df['iyr_close'] = iyr_close
#df['bbh_close'] = bbh_close
```

```python
df.columns
```

```python
x= df[['xlf_close', 'xle_close', 'oih_close', 'xlk_close', 'ige_close',
       'pho_close', 'xlp_close', 'xlu_close', 'moo_close',
       'gdx_close', 'pph_close', 'pbw_close', 'iyr_close']]
```

```python
x = x.pct_change(1)
```

```python
len(x)
```

```python
y = pd.read_csv('SPY.csv')['Adj Close']
```

```python
len(y)
```

```python
y = y.pct_change(1)
```

```python
y.head()
```

```python
y = y.dropna(axis =0, how ='any')
```

```python
len(y)
```

```python
x.head()
```

```python
x = x.dropna(axis =0, how ='any')
```

```python
len(x)
```

```python
import scipy
nnls = scipy.optimize.nnls(x, y)
```

```python
#weights
nnls[0]
```

```python
sectorweig = pd.DataFrame()
sectorweig['sector'] = list(x.columns.values)
sectorweig['weight'] = nnls[0]
```

```python
sectorweig.sort_values(by ='weight',ascending = False)
```

# According to the weights above, we know xlk(Technology Select Sector SPDR ETF ) ranks first with 36.78%, then xlp(Consumer Staples Select Sector SPDR ETF) ranks second with 15.77%, then pph(VanEck Vectors Pharmaceutical ETF) ranks third with 11.14%, then pho(PowerShares Water Resources ETF) ranks fourth with 10.11%. 

```python
#residual
nnls[1]
```

```python
#Approximated Return by Top 13 ETF sectors
res = np.dot(x,nnls[0])
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize = (10,10))
plt.plot(x.index,res,label = 'Approximated Return by Top 13 ETF sectors')
plt.plot(x.index,y,label = 'SPY Return')
plt.ylabel('Adjusted Return')
plt.title('Regress SPY return with top 13 ETF sectors')
plt.legend()
plt.show()
```
