```python
def pca_anal(df_ret,corr_mat, df_std):
    #eigenvalues
    eig_vals = np.linalg.eig(corr_mat)[0]
    #eigenvectors
    eig_vecs = np.linalg.eig(corr_mat)[1]

    #the first eigenvector
    eigenvec_1 = eig_vecs[:,0]
    w1 = eigenvec_1/df_std *(1/math.sqrt(eig_vals[0]))
    theta1 = w1/sum(w1)
    # replace all null values in adjusted return data with 0
    df_ret =df_ret.fillna(0)
    df_ret = df_ret.astype(float)
    df_ret_mat = df_ret.as_matrix()
    dollar_one_norm_ret = np.matmul(df_ret_mat[1:],theta1)
    ret = np.cumprod(dollar_one_norm_ret+1)
    return ret[-1], theta1,df_ret_mat, ret
    
```

```python
def cal_ret(df_ret_mat, weights):
    dollar_one_norm_ret = np.matmul(df_ret_mat[1:],weights)
    ret = np.cumprod(dollar_one_norm_ret+1)
    return ret    
```
