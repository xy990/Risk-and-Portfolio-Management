# 1. Prove Kolmogorov-Smirnoff Theorem: the empirical distribution function converges uniformly to the true distribution.
$lim_{n\to\infty} \sup_{x}|F_N(x)-F(x)| = 0$ 


Let $X_1,..., X_n$ be independent and indentically distributed random variables, with distribution function $F(x) = P({X_1\leq x})$. The empirical cumulative distribution function, also know simply as the empirical distribution function, is defined as $F_n(x) = \frac{1}{n}\sum_{i=1}^{n}\mathbf1\{X_i\leq x\}$, where $\mathbf1$ is the indicator  function, namely $\mathbf1\{X_i\leq x\}$ is one if $X_i\leq x$ and zero otherwise. 
Let $\epsilon >0$. Then fix $ k > \frac{1}{\epsilon}$, and then consider 'knot' points $k_0, ..., k_k$ such that  $-\infty = k_0 < k_1 \leq k_2 \leq ... \leq k_k-1 < k_k = \infty$ that define a partition of $\mathbb{R}$ into k disjoint intervals such that $F(k_{j}^{-}) \leq \frac{j}{k} \leq F(k_j)  j =1, ..., k-1$
where, for each j, $ F(k_{j}^{-}) = P[X_j < k_j] = F(k_j) - P[X = k_j]$
Then, by constructio, if $k_j-1 <k_j$,
$F(k_{j}^{-}) - F(k_j-1) \leq \frac{j}{k} - \frac{j-1}{k} = \frac{1}{k} < \epsilon$
Recall in the following that $ F_n(x)$ is a randome quantity. Now, by the strong law of large numbers, we have pointwise convergence, so that, as $ n \rightarrow \infty$, for $ j =1, ... , k-1$.  
$ F_n(k_j) \xrightarrow{a.s.} F(k_j)$ and  $ F_n(k_{j}^{-}) \xrightarrow{a.s.} F(k_{j}^{-})$.
Then it immediately follows that, for each j, $|F_n(k_{j}) - F(k_{j})| \xrightarrow{a.s.}0$ and  $|F_n(k_{j}^{-}) - F(k_{j}^{-})| \xrightarrow{a.s.}0$   as $n \rightarrow \infty$, so looking at the maximum over all j,  $\triangle_{n} = \max\limits_{j =1,..., k-1}\{|F_n(k_j) - F(k_j)|, |F_n(k_{j}^{-}) -F(k_{j}^{-})|\} \xrightarrow{a.s.} 0$ as $n \rightarrow \infty$.
For any $x$, find the interval within which $x$ lies, that is, identify $j$ such that $k_{j-1} \leq x < k_j$.
Then we have $ F_n(x) -F(x) \leq F_n(k_{j}^{-}) - F(k_{j-1}) \leq F_n(k_{j}^{-}) - F(k_{j}^{-}) + \epsilon$
and $F_n(x) -F(x) \geq F_n(k_{j-1}) - F(k_{j}^{-}) \geq F_n(k_{j-1}) - F(k_{j-1}) -\epsilon$
and thus for any x, $F_n(k_{j-1}) -F(k_{j-1}) -\epsilon \leq F_n(x) -F(x) \leq F_n(k_{j}^{-}) - F(k_{j}^{-}) + \epsilon$
and thus $|F_n(x) - F(x)| \leq \triangle_{n} + \epsilon \xrightarrow{a.s.} \epsilon $ as $ n \rightarrow \infty$.
Hence,as this holds for arbitary x,it follows that $\sup_{x \in \mathbb{R}} \xrightarrow{a.s.} \epsilon$ as $n \rightarrow \infty$ which holds for every $\epsilon >0$.
That is, $\lim_{n\to\infty} \sup_{x\in \mathbb{R}}|F_N(x)-F(x)| = 0$



# 2. Facebook stock and bit coin stock tails the qqplot

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
def normalized_daily_return(x):
    df =pd.read_csv('C:/Users/sherryyang/Desktop/Risk/assignment2/'+ x)
    df_price = pd.DataFrame(df['Adj Close'])
    daily_return = df_price.pct_change(1)
    daily_return = daily_return.dropna()
    daily_ret_normlized = (daily_return-daily_return.mean())/daily_return.std()
    #daily_ret_normlized
    daily_ret_neg = -daily_ret_normlized
    daily_return_sym = pd.concat([daily_ret_normlized, daily_ret_neg], ignore_index=True)
    daily_return_sym = daily_return_sym.values.flatten()
    daily_ret_sym_orderd = np.sort(daily_return_sym)
    return daily_ret_sym_orderd
    
    
```

```python
fb = normalized_daily_return('FB.csv')
```

```python
fb[:10]
```

```python
fb[-10:]
```

```python
 """
 Methods: Plot the qqplot about the daily return of facebook stocks and data generated by student-t distribution( Use percent point function (inverse of cdf  percentiles))
And then use qqplot to plot the two probplots.

 """
    
for df in range(1,21):
    tps = []
    n = len(fb)+1
    for i in range(1, n+1):
        tp= stats.t.ppf((i/n), df, loc=0, scale=1)
        tps.append(tp)
    tps = np.array(tps)
    tps = tps[:-1]
    #fb = fb[:-1]

    #pp_fb = sm.ProbPlot(fb)
    #pp_fb_t = sm.ProbPlot(tps)
    #fig = sm.qqplot_2samples(pp_fb, pp_fb_t, line ='q')
    fig = sm.qqplot_2samples(fb, tps, line ='q')
    fig.suptitle('Degree of Freedom'+ str(df), fontsize=14, fontweight='bold')
    plt.show()
# Student-t test with freedom degree 3 fits the data best, heavy tailed, linearly fit.  
```

"""
 Methods2 :qqplot of the daily return of facebook against quantiles of t-distribution

"""
for df in range(1,21):
    fig = sm.qqplot(fb, stats.t, distargs=(df,), loc=0, scale=1, line ='q')

    plt.show()

```python
"""
Methods 1: use scipy.stats.anderson_ksamp to calculate the anderson darling statistic
"""
from scipy import stats
res = []
for df in range(1,30):
    tps = []
    n = len(fb)+1
    # ppf of student-t distribution
    for i in range(1, n+1):
        tp= stats.t.ppf((i/n), df, loc=0, scale=1)
        tps.append(tp)
    tps = np.array(tps)
    tps = tps[:-1]
    
    res.append((df,stats.anderson_ksamp([fb,tps])[0]))
    print('With Degree of Freedom ' + str(df)+', ' + str(stats.anderson_ksamp([fb, tps])))
res.sort(key=lambda tup: tup[1])
print(res)
```

```python
"""
Calculate the anderson darling statistic using the formula
"""


statistics = []
for df in range(1,21):
    res = 0
    n = len(fb)
    for i in range(1, n+1):
        res += (2*i-1)/n*(np.log(stats.t.cdf(fb[i-1], df, loc=0, scale=1))+np.log(1- stats.t.cdf(-fb[i-1], df, loc=0, scale=1)))
    statistic = -n - res
    statistics.append((df, statistic))
#statistics = statistics.sort(key=lambda tup: tup[1])
#print(statistics)

statistics   
```
