```python
import pandas as pd 
import numpy as np
wmt_close = pd.read_csv('wmt_close.csv',header =1)
```

```python
wmt_close = wmt_close.drop(['Open','High','Low','Volume'],axis =1)
```

```python
wmt_close['Date'] = pd.to_datetime(wmt_close['Date'])
```

```python
wmt_close[:10]
```

```python
wmt_dividend = pd.read_csv('wmt_dividend.csv',header =1)
```

```python
wmt_dividend = wmt_dividend.drop(['Unnamed: 2','Unnamed: 3','Adjusted Close','Close'],axis =1)
wmt_dividend['Date'] = pd.to_datetime(wmt_dividend['Date'])
```

```python
wmt_dividend
```

```python
merge=pd.merge(wmt_close,wmt_dividend, how='outer', left_on='Date', right_on='Date')
```

```python
merge[:10]
```

```python
#The number of closing price is null when merged
sum(np.isnan(merge['Close']))
```

```python
#The number of dividends in this duration
sum(~np.isnan(merge['Dividends']))
```
