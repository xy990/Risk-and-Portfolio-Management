```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
uso = pd.read_csv('USO.csv')
wti = pd.read_csv('DCOILWTICO.csv')
```

```python
uso = uso[['Date', 'Adj Close']]
```

```python
uso =uso.set_index('Date')
```

```python
wti = wti.set_index('DATE')
```

```python
uso.head()
```

```python
wti.head()
```

```python
len(wti),len(uso)
```

```python
df = pd.merge(wti, uso, left_index=True, right_index=True)
```

```python
df = df[df['DCOILWTICO'] != '.']
```

```python
df.columns = ['wti', 'uso']
```

```python
len(df)
```

```python
df = df.astype(float)
```

```python
df['wti_daily'] = df['wti'].pct_change(1)
df['uso_daily'] = df['uso'].pct_change(1)
```

```python
wti_ret = np.zeros(len(df))
uso_ret = np.zeros(len(df))
for i in range(len(df)):
    wti_ret[i] = np.prod((df['wti_daily']+1)[:i])-1
    uso_ret[i] = np.prod((df['uso_daily']+1)[:i])-1
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize = (20,10))
plt.plot(df.index, uso_ret, label ='uso', color ='r')
plt.plot(df.index,wti_ret,label = 'wti')
plt.ylabel('Return')
plt.title('The return of USO and WTI')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize = (20,10))
plt.plot(df.index, df['uso'], label ='uso', color ='r')
plt.plot(df.index,df['wti'],label = 'wti')
plt.ylabel('Price')
plt.title('The Price of USO and WTI')
plt.legend()
plt.show()
```

# 2a Simulate short-short 

```python
fas = pd.read_csv('FAS.csv')[['Date','Adj Close']]
faz = pd.read_csv('FAZ.csv')[['Date','Adj Close']]
uyg = pd.read_csv('UYG.csv')[['Date','Adj Close']]
skf = pd.read_csv('SKF.csv')[['Date','Adj Close']]
edc = pd.read_csv('EDC.csv')[['Date','Adj Close']]
edz = pd.read_csv('EDZ.csv')[['Date','Adj Close']]
```

```python
fas = fas.set_index('Date')
faz = faz.set_index('Date')
uyg = uyg.set_index('Date')
skf = skf.set_index('Date')
edc = edc.set_index('Date')
edz = edz.set_index('Date')
```

```python
da1 = pd.merge(fas, faz, left_index=True, right_index=True)
da1.columns = ['fas', 'faz']
```

```python
da2 = pd.merge(uyg, skf, left_index=True, right_index=True)
da2.columns = ['uyg', 'skf']
```

```python
da3 = pd.merge(edc, edz, left_index=True, right_index=True)
da3.columns = ['edc', 'edz']
```

```python
da4 = pd.merge(da1, da2, left_index=True, right_index=True)
da5 = pd.merge(da4, da3, left_index=True, right_index=True)
```

```python
da5.head()
```

```python
len(da5)
```

```python
da5['fas_daily'] = da5['fas'].pct_change(1)
da5['faz_daily'] = da5['faz'].pct_change(1)
da5['uyg_daily'] = da5['uyg'].pct_change(1)
da5['skf_daily'] = da5['skf'].pct_change(1)
da5['edc_daily'] = da5['edc'].pct_change(1)
da5['edz_daily'] = da5['edz'].pct_change(1)
```

```python
da5.head()
```

```python
da5= da5.dropna(axis =0, how ='any')
```

```python
#There is split on 2018-03-28, so i change the return to 0 to indicate no change
da5.ix[da5.index == '2018-03-28', 'edz_daily'] =0
```

# Without borrowing cost

```python
da5['fas_faz'] = 1-da5['fas_daily'] - da5['faz_daily']
da5['uyg_skf'] = 1-da5['uyg_daily'] - da5['skf_daily']
da5['edc_edz'] = 1-da5['edc_daily'] - da5['edz_daily']
```

```python
pair1_ret = np.zeros(len(da5))
pair2_ret = np.zeros(len(da5))
pair3_ret = np.zeros(len(da5))
for i in range(len(da5)):
    pair1_ret[i] = np.prod((da5['fas_faz'])[:i])-1
    pair2_ret[i] = np.prod((da5['uyg_skf'])[:i])-1
    pair3_ret[i] = np.prod((da5['edc_edz'])[:i])-1
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize = (20,10))
plt.plot(da5.index, pair1_ret, label ='fas_faz')
plt.plot(da5.index, pair2_ret,label = 'uyg_skf')
plt.plot(da5.index, pair3_ret,label = 'edc_edz')
plt.ylabel('Return')
plt.title('Short-Short Without Borrowing Cost')
plt.legend()
plt.show()
```

# With borrowing cost

```python
fas_fee = 2.1824
faz_fee = 12.7218
uyg_fee = 4.6235
skf_fee =9.6865
edc_fee = 6.9736
edz_fee = 12.9827
```

```python
da5['fas_faz_cost'] = 1-da5['fas_daily'] - da5['faz_daily'] -  fas_fee*0.01/252 - faz_fee*0.01/252
da5['uyg_skf_cost'] = 1-da5['uyg_daily'] - da5['skf_daily'] - uyg_fee*0.01/252 - skf_fee*0.01/252
da5['edc_edz_cost'] = 1-da5['edc_daily'] - da5['edz_daily'] - edc_fee *0.01/252 - edz_fee *0.01/252
```

```python
da5.head()
```

```python
pair1_ret_cost = np.zeros(len(da5))
pair2_ret_cost = np.zeros(len(da5))
pair3_ret_cost = np.zeros(len(da5))
for i in range(len(da5)):
    pair1_ret_cost[i] = np.prod((da5['fas_faz_cost'])[:i])-1
    pair2_ret_cost[i] = np.prod((da5['uyg_skf_cost'])[:i])-1
    pair3_ret_cost[i] = np.prod((da5['edc_edz_cost'])[:i])-1
```

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots(figsize = (20,10))
plt.plot(da5.index, pair1_ret_cost, label ='fas_faz')
plt.plot(da5.index, pair2_ret_cost,label = 'uyg_skf')
plt.plot(da5.index, pair3_ret_cost,label = 'edc_edz')
plt.ylabel('Return')
plt.title('Short-Short With Borrowing Cost')
plt.legend()
plt.show()
```
